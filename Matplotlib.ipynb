### This will serve as refresher on Matplotlib and Basic Visualizations
* Matplotlib is a python library used for data visualization
* Can create bar-plots, scatter-plots, histograms and a lot more with matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.arange(1,11)
y1 = 2 * x
y2 = 3 *x
```

```python
#Make Plot
plt.plot(x,y1, color = 'red', linewidth = 4)
plt.plot(x, y2, color =  'black', linewidth =3, linestyle = ':')
#Add Title
plt.title('First Line Graph')
#Add x & y labels
plt.xlabel('Happiness')
plt.ylabel('Hurricanes Wins')
plt.grid()
plt.show()
```

```python
#bar-plots\
shire = {"Sam": 40, "Frodo": 90, "Merry": 60, "Pippen": 70}
names =  list(shire.keys())
impact = list(shire.values())

names, impact
plt.barh(names, impact, color = 'gold')
plt.title('Likeability of Hobbits')
plt.ylabel('Hobbits')
plt.xlabel('Scores')
#plt.grid()
plt.show()
```

```python
#Scatter plots
x = [4,7,3,9,8,5]
y = [3,7,6,8,4,5]
y2 = [7,5,6,8,4,6]

plt.scatter(x,y)
plt.scatter(x,y2, color = 'orange')
plt.grid()
```

```python
#Histograms gives us the distribution of continuous columns
iris = pd.read_csv('C:/Users/locone/Documents/2020/Python/Prac_data/iris.csv')
print(iris.columns)
plt.hist(iris['sepal.length'], bins = 20, color = 'pink')
plt.title('Distribution of Sepal Length')
plt.xlabel('Sepal Length cm')
plt.show()
```

```python
#box plots how does one variable change
#iris.boxplot(column = 'sepal.width', by = 'variety')

#Use seaborn when possible it makes plots look a litte bit better
import seaborn as sns
sns.boxplot(x= iris['variety'], y = iris['sepal.length'])
```

```python
#Pie Charts are the final plot
food = ['bread', 'beer', 'pizza', 'tacos']
cost = [30, 80, 45, 28]

#This will change the size of the plot
plt.figure(figsize = (10,10))

plt.pie(cost, labels =food, autopct = '%0.1f%%', shadow = True)
plt.title('Cost breakdown per month')
plt.show()
```
